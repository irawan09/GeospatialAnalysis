## 2 - Reading CSV with Pandas and Converting to GeoDataFrame

Comma-separated values (CSV)are predominantly the most used files in data science. Many datasets, however, have geographic component (Latitude/Longitude) that could be easily converted into geometry and could further be used for spatial data analysis.

### Learning Objectives:

1. Read csv files and convert into Geopandas GeodataFrame.
2. Writing Data into Geospatial Format

```python
import geopandas as gpd
import pandas as pd
```

```python
df = pd.read_csv("data/acled2019.csv")
df.head()
```

```python
df.shape
```

```python
df.sample(100).plot()
```

* We use geopandas points_from_xy() to transform Longitude and Latitude into a list of shapely.Point objects and set it as a geometry while creating the GeoDataFrame.

* What we are interested in is the latitude and longitude columns of the dataset. 

```python
gdf = gpd.GeoDataFrame(
    df, 
    geometry=gpd.points_from_xy(df.longitude, df.latitude)
)
gdf.head()
```

```python
gdf.plot(markersize=2);
```

## Writing Data to Spatial Format.

It is not ideal to read your file and then convert it all the time. This will take time and computatin resources. So, let us convert the CSV data to Spatial format. This might take a while depending on computing resources and the size of the data.

* Write to Shapefile 

```python
gdf.to_file("output/acled2019.shp", crs={'init' :'epsg:4326'})
```

* Write to GeoJSON

```python
gdf.sample(1000).to_file("output/acled2019-Geojson.geojson", driver='GeoJSON')
```

**Recap**
* `points_from_xy()` --> convert latitude & longitude into Geometry (Geodataframe)
* `to_file()` --> writes data into spatial format (Shapefile, GeoJSON, GeoPackage)
