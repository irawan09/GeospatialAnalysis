## 5 - CRS & Projections

Coordinate reference systems represent how our data as two dimensional (planar) relates to actual places on earth. It is the glue that holds the attributes to their respective locations. Geodataframes has .crs attribute that can give you the original CRS used in the data. Projected Coordinate systems, on the other hand, tells the data how to draw on a flat surface, like on a paper map or a computer screen


It is easy to transform and project these coordinates. However, to perform projections, it is necessary to have the same CRS  in order to carry out geographic analysis and get the right values out the analysis. The countries, cities and rivers have the same CRS. Let us check them.

### Learning Objectives:

1. Learn Projected Coordinate Reference Systems.


```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Read Countries data
countries = gpd.read_file('data/countries.shp')
countries.head()
```

```python
countries.crs
```

```python
# Read Conflict data
conflicts = gpd.read_file("output/acled2019.shp", rows=500)
conflicts.head()
```

```python
conflicts.crs
```

```python
conflicts.crs.name
```

Different CRS have different measurements. For some, the coordinates are defined in decimal degrees while others are defined in meters. It is common process to reproject data from one format to another in Geographic data processing. 

Resource for getting the rigt projection 
https://spatialreference.org/

We will project our data into mercator. The Mercator projection(3395), latitude-longitude quadrangles are stretched along the x-axis and y-axis as you move away from the equator.

```python
# Look at the geometry column: decimal degrees
countries.geometry[:5]
```

```python
countries = countries[(countries.NAME != "Antarctica") & (countries.NAME != "Fr. S. Antarctic Lands")]
# Project the data into Mercator Projection epsg=3395

countries_projected = countries.to_crs(epsg=3395)

# See the geometry column of the projected countries
countries_projected['geometry'][:5]
```

Due to the projection, the geometry is no longer in lat/lon style. It is easier to understand the difference in maps. Let us plot both the original countries and the projected countries.

```python
# Let us plot both unprojected and projected countries
countries.plot();
countries_projected.plot();
```

Notice the different scales of x and y in both maps. If you try to overlay the projected data with unprojected data, then your data will not align properly. Let us see if we can plot conflicts on the top of projected countries. Remember we have not projected the conflict data.

```python
fig, ax = plt.subplots()
countries_projected.plot(ax=ax)
conflicts.plot(ax=ax, color='red') # Cities are still in WGS84
```

```python
fig, ax = plt.subplots()
countries.plot(ax=ax) # countries are in WGS84
conflicts.plot(ax=ax, color='red') # conflicts are in WGS84
```

```python
conflicts_projected = conflicts.to_crs(epsg=3395)
fig, ax = plt.subplots()
countries_projected.plot(ax=ax)
conflicts_projected.plot(ax=ax, color='red')

```

**Recap**
* `.crs()` --> Displays information about Coordinate reference system of the data.
* `.to_crs()` --> projects data from one coordinate to another.
