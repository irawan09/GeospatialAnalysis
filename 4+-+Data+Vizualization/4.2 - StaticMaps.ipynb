## 4.2 - Static Maps

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

## Polygon

```python
small_areas = gpd.read_file("data/malmo-pop.shp")
small_areas.head()
```

```python
small_areas.shape
```

```python
small_areas.crs
```

```python
small_areas.plot();
```

```python
small_areas.plot(figsize=(12,10));
```

```python
small_areas.plot(figsize=(12,10),
                color="gray");
```

```python
small_areas.plot(figsize=(12,10),
                color="gray",
                alpha=0.7);
```

```python
small_areas.plot(figsize=(12,10),
                color="gray",
                alpha=0.7,
                edgecolor="#CFAC84");
```

```python
small_areas.plot(figsize=(12,10),
                color="gray",
                alpha=0.7,
                edgecolor="#CFAC84",
                linewidth=2);
```
