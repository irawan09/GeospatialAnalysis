## 4.3 - Multi Layer Maps

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
small_areas = gpd.read_file("data/malmo-pop.shp")
roads = gpd.read_file("data/malmo-streets.shp")
preschools = gpd.read_file("data/preschools.shp")
```

### Side by Side Plots

```python
fig, ax  = plt.subplots(3, figsize=(14,12))
```

```python
fig, ax  = plt.subplots(1, 3, figsize=(14,12))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20,18))
small_areas.plot(ax=ax[0], figsize=(18,14))
roads.plot(ax=ax[1], figsize=(18,14))
preschools.plot(ax=ax[2], figsize=(18,14))
```

### Overlay Plots

```python
fig, ax = plt.subplots(1,1,  figsize=(12,10))
roads.plot(ax=ax, color='#C05A3C', linewidth=0.3, alpha=0.5);
small_areas.plot(ax= ax,
                color='#F1ECEB',
                alpha=0.7);
preschools.plot(ax = ax,
                figsize=(12,12),
                markersize=40,
               color="black",
               edgecolor="white",
               alpha=0.8,
               marker="o");


plt.title('Mlmo Neighbourhoods', fontsize=40, fontname="Palatino Linotype", color="grey")
ax.axis("off")
plt.axis('equal')
plt.show()
```
